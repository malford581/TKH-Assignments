```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exercise 1

Plot one chart using matplotlib (OOP-style) using the following data.

```python
df = pd.read_csv('../datasets/GOOG.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.pairplot(df)
```

```python
df1= df["Date"][0:15]
df2 = df["High"][0:15]
df3 = df["Low"][0:15]
```

```python
fig, ax = plt.subplots(figsize=(20, 15))
ax.plot(df1, df2, label='High') 
ax.plot(df1, df3, label='Low')  
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[0, 10], xlabel='Date', ylabel='Values',
       title="Google's High and Lows for 15 day period")
ax.legend()
plt.show()
```

Perform some initial exploratory analysis to the following dataframe. Then plot two or more charts related to the data.

```python
df.info()
```

```python
fig, ax = plt.subplots(figsize=(10, 20))
ax.barh(df["School name"], df['Number of students who registered for the SHSAT'])
ax.set(xlim = [0, 150], xlabel='Number of registered students', ylabel='Name of School',
       title='Total Students who registered for SHSAT')
```

```python
df
```

```python
sns.pairplot(df)
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.jointplot(x=df['Number of students who registered for the SHSAT'].values, y=df['Number of students who took the SHSAT'].values, height=10)
plt.xlabel('Number of students who registered for the SHSAT')
plt.ylabel('Number of students who took the SHSAT')
plt.show()
```
